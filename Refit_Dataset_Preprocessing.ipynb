```python
# Write to a CSV file
import csv
field_names= ['id', 'startDateTime', 'occupancyType', 'builtFormType', 'orientation', 'wallTypeMainBuilding', 'wallAgeBandMainBuilding', 'wallTypeExtension', 'wallAgeBandExtension', 'cavityWallInsulationPresent', 'windowType', 'loftType', 'loftInsulationType', 'loftInsulationThickness', 'id1', 'startDateTime', 'applianceType', 'plugIdRef', 'spaceIdRef', 'id2', 'variableType', 'units', 'shape' ]
```

```python
list = []

for building in root1.findall('{http://www.refitsmarthomes.org}Building'):
    #print(building.attrib.items())
    #for key, value in building.attrib.items():
    #  print(key, ' -> ', value)
    print('--------------------------------------------------------------------------')
    dict1 = building.attrib
    print(building.attrib)
    print('--------------------------------------------------------------------------')
    for child in building:
      #print(child.tag, child.attrib)
      if child.tag.endswith('Appliance'):
        dict = child.attrib
        #print(dict)
        #print(dict['applianceType'])
        if(dict['applianceType'] in ['Dehumidifier/heater', 'Dishwasher', 'Electric heater', 'Fridge', 'Freezer', 'Fridge freezer', 'Hair dryer', 'Kettle', 'Oven', 'Microwave', 'Television', 'Telephone', 'Toaster', 'Tumble dryer', 'Washing machine', 'Washer dryer']):
          plugref = dict['plugIdRef']
          print(dict['applianceType'], plugref)
          #print(child.get('id'=dict['plugIdRef']))
          dict['id1'] = dict['id']
          del dict['id']
          resDict = {**dict1, **dict}
          for child1 in building.findall('{http://www.refitsmarthomes.org}Plug'):
            #print(child1.attrib)
            if child1.attrib['id'] == plugref:
              print(child1.attrib)
              for child2 in child1:
                #print(child2.attrib)
                #print(child2[0].attrib)
                timeseriesvalue = child2[0].attrib['id']
                resDict['id2'] = child2[0].attrib['id']
                resDict['variableType'] = child2[0].attrib['variableType']
                resDict['units'] = child2[0].attrib['units']
                print(timeseriesvalue)
                df = timeseriespd.loc[timeseriespd['TimeSeriesVariable/@id'] == timeseriesvalue]
                print(df.shape)
                resDict['shape'] = df.shape[0]
          print('***********************************************')
          print(resDict)
          list.append(resDict)
          print('***********************************************')
```

```python
print(list)
```

```python
with open('Preprocessed_Refit.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(list)
```
